#  Train from scratch

```python
import uuid   # Unique identifier
import os
import time
```

```python
# Define paths and labels
IMAGES_PATH = os.path.join('data', 'images')
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
# Initialize webcam capture
cap = cv2.VideoCapture(0)

# Loop through each label
for label in labels:
    print('Collecting images for {}'.format(label))
    
    # Give a few seconds to prepare for image capture
    time.sleep(5)
    
    # Loop through the specified number of images
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        # Capture a frame from the webcam
        ret, frame = cap.read()
        
        # Construct a unique image name based on label and UUID
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        
        # Save the captured frame as an image file
        cv2.imwrite(imgname, frame)
        
        # Display the captured frame in a window
        cv2.imshow('Image Collection', frame)
        
        # Wait for 2 seconds between captures
        time.sleep(2)
        
        # Break the loop if 'q' key is pressed
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Release the webcam capture and close windows
cap.release()
cv2.destroyAllWindows()
```

# Live Testing

```python
# Open the webcam
cap = cv2.VideoCapture(0)

# Loop to continuously capture frames from the webcam
while cap.isOpened():
    # Read a frame from the webcam
    ret, frame = cap.read()
    
    # Make detections using the YOLO model
    results = model(frame)
    
    # Display the frame with bounding boxes and labels using OpenCV
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    # Check if the 'q' key is pressed to exit the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
